### Q1: 小明的家庭作業裏有很多數列填空練習。 
說明：
填空練習的要求是：已知數列的前四項，填出第五項。 
因為已經知道這些數列只可能是等差或等比數列，他決定寫一個程式來完成這些練習 

<b>輸入說明: <br></b>
第一行是數列的數目t（1 <= t <= 5）。<br> 
以下每行均包含四個整數，表示數列的前四項。 <br>

<b>輸出說明: <br></b>
對輸入的每個數列，輸出它的前五項。<br>
不合法的輸入則輸出E。 <br>
<blockquote>
input : <br>
2 <br>
1 2 3 4 <br>
32 16 8 4 <br>

output: <br>
1 2 3 4 5 <br>
32 16 8 4 2<br>
</blockquote>

```python
try:
    t = int(input())
    for i in range(t):
        a, b, c, d= map(int, input().split())
        if b / a == c / b and b / a == d / c:
            if (d * c) / b >= 1: 
                print("{} {} {} {} {}".format(a, b, c, d, (d * c) // b))
            else: #處理第五項可能為小於一之小數的情況
                print("{} {} {} {} {}".format(a, b, c, d, (d * c) / b))
        elif b - a == c - b and b - a == d - c:
            print("{} {} {} {} {}".format(a, b, c, d, d + c - b))
        else:
            raise Exception
except:
    print("E")
```

### Q2: 請訂出list: name_input (包含三個英文名 Peter, Nina, Annie)
說明：
dict: names (包含兩個英文名 Peter, Nina), classes (包含兩科目 Math, English)
用for 迴圈 配合if...elif...else 判斷name_input並一次印出以下結果
<blockquote>
Peter likes Math<br>
Nina likes English<br>
This name is not in name dictionary!
</blockquote>

```python
mydict = {"Peter" : "Math", "Nina" : "English"}
for i in range(3):
    name = input()
    if all(x != name for x in mydict):
        print("This name is not in name dictionary!")
    else:
        print("{} likes {}".format(name, mydict.get(name)))
```

### Q3: 小明修了很多不同老師的課程，他想知道自己在哪個老師的課程中，平均得到幾分來決定下學期要修哪個老師的課程。 
<b>輸入說明<br></b>
(1)第一筆輸入為輸入N M，N為小明所修課程的數量，M為老師的數量，2小於N小於100、0小於M小於10。 <br>
(2)接著輸入N筆課程的課程編號、課程名稱及期末總分，以空格隔開課程名稱及期末總分。請使用"字典"儲存。 <br>
(3)接著輸入M位老師的老師編號、老師名字以及課程編號，以空格隔開老師編號、老師名字以及課程編號。請使用串列儲存。 <br>
(4)輸入要查詢的老師名字。 <br>

--------------------------------------------------------------------------------------------------------------------------- 
<b>輸出說明 <br></b>
輸出平均分數(整數)。若第一筆輸入不在範圍輸出"Error"，若找不到老師名字請輸出"查無此人"。 <br>

--------------------------------------------------------------------------------------------------------------------------- 
<blockquote>
Sample Input:<br>
5 3 <br>
82432 計算機概論與程式設計上 90 <br>
42785 邏輯電路 75 <br>
87476 計算機概論與程式設計下 96 <br>
64042 中等流體力學 65 <br>
35395 電子電路設計 60 <br>
001 黃乾綱 82432 87476 <br>
002 吳文中 42785 35395 <br>
003 蔡武廷 64042 <br>
黃乾綱 <br>

Sample Output:<br>
93 <br>
</blockquote>

```python
n, m = map(int, input().split())
final_score = 0
teachername = []
t_c = []
mydict = {}
for i in range(n):
    classno, classname, score = map(str, input().split())
    mydict[classno] = int(score)
for i in range(m):
    a, b, c = map(str, input().split(" ", 2))
    teachername.append(b)
    t_c.append(c)
namereq = input()
flag = 0
for i in range(m):
    if namereq == teachername[i]:
        flag = 1
        break
if flag == 0:
    print("查無此人")
else:
    classes = list(map(str, t_c[i].split()))
    for i in range(len(classes)):
        n = mydict.get(classes[i])
        final_score += n
    print("{}".format(final_score // len(classes)))
```
